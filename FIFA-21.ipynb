```python
#value
def convert_currency(value):
    if 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1000
    elif 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1000000
    else:
        return float(value.replace('€', ''))

# apply the function to the 'Value' column
df['value'] = df['value'].apply(convert_currency)

df
```

```python
df['value'].dtypes
```

```python
#wage
def convert_currency(wage):
    if 'K' in wage:
        return float(wage.replace('€', '').replace('K', '')) * 1000
    else:
        return float(wage.replace('€', ''))

# apply the function to the 'wage' column
df['wage'] = df['wage'].apply(convert_currency)

df
```

```python
# hits
def convert_currency(hits):
    if 'K' in hits:
        return float(hits.replace('K', '')) * 1000
    else:
        return float()

# apply the function to the 'wage' column
df['hits'] = df['hits'].apply(convert_currency)
```

```python
# release_clause
def convert_currency(release_clause):
    if 'K' in release_clause:
        return float(release_clause.replace('€', '').replace('K', '')) * 1000
    elif 'M' in release_clause:
        return float(release_clause.replace('€', '').replace('M', '')) * 1000000
    else:
        return float(release_clause.replace('€', ''))

# apply the function to the 'Value' column
df['release_clause'] = df['release_clause'].apply(convert_currency)

df
```

```python
X=df.drop('ova', axis=1)
y=df['ova']
y
```

```python
X_num=X._get_numeric_data()
X_num
```

```python
X_cat1=pd.get_dummies(X_cat1, drop_first=True)
X=pd.concat([X_num, X_cat1], axis=1) # concat numerical and categorical transformations
X
```

```python
from sklearn.preprocessing import MinMaxScaler
MinMaxtransformer = MinMaxScaler().fit(X)      #no refit!!
X1 = MinMaxtransformer.transform(X)
print(X1.shape)
X1 = pd.DataFrame(X1,columns=X.columns)
X1.head()
```

```python
X1.describe().T
```

Train-test split.

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, test_size=0.2, random_state=7) 
```

```python
print(X1_train.shape)
print(X1_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X1_train.head()
```

```python
y_train.head()
```

```python
#train/fit model
lm = linear_model.LinearRegression()
lm.fit(X1_train,y_train)                         #no refit!!
```

```python
from sklearn.metrics import r2_score

predictions = lm.predict(X1_train)
r2_score(y_train, predictions)
```

```python
predictions_test = lm.predict(X1_test)
r2_score(y_test, predictions_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
#the mean of the squared differences between the predicted and actual values. 
#MSE = (1/n) * Σ(yi - ŷi)^2
#A lower MSE indicates a better fit between the model and the data.
```
